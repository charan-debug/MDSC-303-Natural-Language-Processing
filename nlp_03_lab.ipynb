```python
import re
fileList= !ls sss
data= ''

#fileList = fileList[0].split()
fileList=["sss_01_01.txt", "sss_01_02.txt", "sss_01_03.txt", "sss_01_04.txt", "sss_01_05.txt"]
for f in fileList:
    f = open("sss/"+f, 'r')
    data += f.read()

data = data.lower()
data = re.sub("[.\“”:?!;,()-]","",data)
lines = data.splitlines()
print(lines)
```

 #create a dictionary bigram with key = (wi,wj) and value = bigram probability P(wj|wi)

```python
 #create a dictionary bigram with key = (wi,wj) and value = bigram probability P(wj|wi)
#for each sentenece we need to add <s> at starting and </s> at ending of a line
lines = data.splitlines()
for i in range(len(lines)):
    lines[i] = "<s> " + lines[i] + " </s>"

print(lines)
```

```python
lines
```

```python
#for each sentence creating a bigram dictionary
bigram = {}
for x in lines:
  words = x.split()
  for i in range(len(words)-1):
    if (words[i], words[i+1]) not in bigram:
      bigram[(words[i], words[i+1])] = 0
    bigram[(words[i], words[i+1])] += 1

#print(bigram_dict)
```

```python
bigram
```

```python
for i in bigram:
  bigram[i]=bigram[i] / len(lines)
bigram
```

```python
# Calculate probabilities
for key in bigram:
  count_w = sum(1 for x in lines for word in x.split() if word == key[0])

  bigram[key] = bigram[key] / count_w
  #print(count_w)
  #print(bigram[key])
#print(bigram)
```

#create a dictionary trigram with key = (wi,wj,wk) and value = trigram probability P(wk|wi wj)


```python
lines = data.splitlines()
for i in range(len(lines)):
    lines[i] = "<s> <s> " + lines[i] + " </s> </s>"

print(lines)
```

```python
#create a dictionary trigram with key = (wi,wj,wk) and value = trigram probability P(wk|wi wj)

trigram = {}
for line in lines:
  words = line.split()
  for i in range(len(words)-2):
    if (words[i], words[i+1], words[i+2]) not in trigram:
      trigram[(words[i], words[i+1], words[i+2])] = 0
    trigram[(words[i], words[i+1], words[i+2])] += 1
```

```python
for i in trigram:
  trigram[i]=trigram[i] / len(lines)
```
