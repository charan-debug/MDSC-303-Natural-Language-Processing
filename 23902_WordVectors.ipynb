```python
#The usual processing of the five discourses to get proper sentences and words
import re
files= ['sss_01_01.txt','sss_01_02.txt','sss_01_03.txt','sss_01_04.txt','sss_01_05.txt']
data= ''
for file in files:
    f= open("sss/"+file, 'r')
    data+= f.read()
data= data.lower()
data= re.sub("[.\“”:?!;,()-]","\n",data)
data= re.sub("(\s[‘’])|([‘’]\s)","\n",data)
data= re.sub("[‘’]s","",data)
data= re.sub("[0-9][A-Za-z]+","",data)
words= data.splitlines()
words= [w.split(' ') for w in words]
words= sum(words,[])
words= [w.strip() for w in words]
words= [w for w in words if len(w) >= 1]
word_set= list(set(words))
word_set.sort()
N= len(words)
modv= len(word_set)
print(N)
print(modv)
```

```python
import numpy as np

# Word to Index, Index to Word
w2i = {}
i2w = {}
for i, w in enumerate(word_set):
    w2i[w] = i
    i2w[i] = w

# Positive samples with window size ws
ps = {}
ns = {}
ws = 5
for i, w in enumerate(words):
  #Q1: Set ls, le, rs and re
    #Your code here...

    ls = max(i - ws, 0)
    le = i
    rs = i + 1
    re = min(i + ws + 1, len(words))

    # Do not change the loops below
    s = set()
    for j in range(ls, le):
        s.add(w2i[words[j]])
    for j in range(rs, re):
        s.add(w2i[words[j]])
    ps[w2i[w]] = s

# Negative samples to be added
k = 2
for i in range(modv):
    added = 0
    s = set()
    while added < k:
        r = np.random.randint(modv)
        if r not in ps[i] and r not in s:
            s.add(r)
            added += 1
    ns[i] = s
```

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

lr = 0.01
embed = 50
w = np.random.randn(modv, embed)  # Initialize word vectors
c = np.random.randn(modv, embed)  # Initialize context vectors
epochs = 50
#sigmoid has been defined above
#np.dot(x,y) takes the dot product between x and y
#np.log(x) takes log of x
#Page 125 and 126 of the Key Text contain the answers to the questions below
for e in range(epochs):
    loss = 0
    for i in range(modv):
        for j in ps[i]:
            dot_product = np.dot(w[i], c[j])
            loss += np.log(sigmoid(dot_product)) #Q3: What is to be added here?
        for j in ns[i]:
            dot_product = np.dot(w[i], c[j])#Q4: What is to be added here?
            loss += np.log(1 - sigmoid(dot_product))
    #Please note the following line before answering Q3 and Q4.
    loss = -loss
    print(loss)

    #Assuming stochastic gradient descent - i.e an update for every sample in the dataset, answer Q5 to Q8.
    for i in range(modv):
        for j in ps[i]:
            dot_product = np.dot(w[i], c[j])
            grad = sigmoid(dot_product) - 1
            c[j] -= lr * grad * w[i]  #Q5: the appropriate grad
            w[i] -= lr * grad * c[j]  #Q6: the appropriate grad
        for j in ns[i]:
            dot_product = np.dot(w[i], c[j])
            grad = sigmoid(dot_product)
            c[j] -= lr * grad * w[i]  #Q7: the appropriate grad
            w[i] -= lr * grad * c[j]  #Q8: the appropriate grad
```
